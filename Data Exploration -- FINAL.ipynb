```python
#import relevant libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Loading the dataframe, and cleaning the lists that were stored as text in the csv file
df = pd.read_csv(r'C:\Users\MainUser\Desktop\Quotes_Cleaned.csv')
df['words'] = df.words.apply(lambda x: x.strip('[]').split(','))
df['words_ex_stopwords'] = df.words_ex_stopwords.apply(lambda x: x.strip('[]').split(','))
df['lemmas'] = df.lemmas.apply(lambda x: x.strip('[]').split(','))
```

```python
# check weightings of classes
df['Author'].value_counts('%').round(2)
```

```python
# examine the sentence length feature
plt.rcParams.update({'font.size': 11})
fig, axes = plt.subplots(figsize=(4.5,5))
sns.boxplot(x="Author", y="sentence_length", data=df)
axes.set_xlabel('Author')
axes.set_ylabel('Sentence Length (# Words)')
axes.set_title('Fig 2: Distribution of Sentence Length')
sns.despine()
```

```python
# examine the sentence length feature closer up
plt.rcParams.update({'font.size': 11})
fig, axes = plt.subplots(figsize=(4.5,5))
sns.boxplot(x="Author", y="sentence_length", data=df)
axes.set_xlabel('Author')
axes.set_ylabel('Sentence Length (# Words)')
axes.set_title('Fig 3: Zoomed Version of Fig 2')
axes.set_ylim(0,30)
sns.despine()
```

```python
# check the really short sentences
df[df['sentence_length'] == 1] 
```

```python
# check the really long sentences
list(df[df['sentence_length'] > 50]['Quote'])
```

```python
# examine the stop words frequency feature
plt.rcParams.update({'font.size': 11})
fig, axes = plt.subplots(figsize=(4.5,5))
sns.boxplot(x="Author", y="stop_word_pct", data=df)
axes.set_xlabel('Author')
axes.set_ylabel('Stop Words (Fraction of Total Words in Sentence)')
axes.set_title('Fig 4: Distribution of Stop Word Frequency')
axes.set_ylim(0,1)
sns.despine()
```
